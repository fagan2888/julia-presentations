# Julia Sets

```julia
using PyPlot
```

```julia
# julia set
# (the familiar mandelbrot set is obtained by setting c==z initially)
function julia(z, c; maxiter=200)
    for n = 1:maxiter
        if abs2(z) > 4
            return n-1
        end
        z = z*z + c
    end
    return maxiter
end
```

```julia
# varying the second argument to julia() tiny amounts results in a stunning variety of forms
@time m = [ uint8(julia(complex(r,i), complex(-.5,.75))) for i=1:-.002:-1, r=-1.5:.002:1.5 ];
```
